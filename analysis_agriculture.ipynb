```python
import pandas as pd
from translation import translate_dataframe
from project.config import api_key
file_path = 'data/indices of agricultural production.csv'
data = pd.read_csv(file_path, delimiter=';', skiprows=5)
```

```python
data.columns = data.iloc[0]  
data = data[1:]  
data = data.reset_index(drop=True)  
data = data.dropna(axis=1, how='all')

data
```

```python
new_column_names = [
    data.iloc[0, 0],  
    data.iloc[2, 1],  
    data.iloc[3, 2],  
    data.iloc[3, 3],  
]

data.columns = new_column_names + data.columns[4:].tolist()
data
```

```python
data = data.iloc[4:]
data = data.iloc[:-2]
data = data.loc[:, data.columns.notna()]
data
```

```python
cleaned_columns = []
for col in list(data.columns):
    if '/' in col:
        cleaned_columns.append(col.split('/')[0].strip())  
    else:
        cleaned_columns.append(col)
print(cleaned_columns)
data.columns = cleaned_columns
data
```

```python
for col in data.columns:
    if col != 'Роки':
        data[col] = data[col].str.replace(',', '.').astype(float)
        
data
```

```python
data['Роки'] = data['Роки'].str.split(' ').str[0]
data
```

```python
data_translated = translate_dataframe(data, api_key)
data_translated
```

```python
import matplotlib.pyplot as plt

data_translated.set_index('Years', inplace=True)

data_translated.plot(figsize=(10, 6), marker = 'o')

plt.xlabel('Years')
plt.ylabel('Values')
plt.title('Progression Over Years of Indices of agricultural production')
plt.legend(title='Columns', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)

plt.tight_layout()
plt.show()
```
